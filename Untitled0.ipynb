```python
import numpy as np
import pandas as pd
#age
columns_to_analyze = ['age', 'Fedu', 'traveltime', 'studytime',
                     'failures', 'famrel', 'freetime', 'Dalc',
                     'absences', 'G2']

for column in columns_to_analyze:
      print(f"\nАнализ выбросов для столбца: {column}")
      Q1 = np.percentile(df[column], 25)
      Q3 = np.percentile(df[column], 75)
      IQR = Q3 - Q1
      lower_bound = Q1 - 1.5 * IQR
      upper_bound = Q3 + 1.5 * IQR

      outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]

      print(f"Границы выбросов: [{lower_bound:.2f}, {upper_bound:.2f}]")
      print(f"Выбросы:\n{outliers.to_string()}")
      print(f"Абсолютное количество выбросов: {len(outliers)}")
      print(f"Относительное количество выбросов: {len(outliers)/len(df)*100:.2f}%")
```

```python
#14) Сколько признаков достаточно для объяснения 90% дисперсии после применения
#метода PCA (предварительно исключите признаки, которые не имеют смысла:
#например, все виды id)?
# 15) Какой признак вносит наибольший вклад в первую компоненту?


from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA()
pca.fit(X_train)
X_pca = pca.transform(X_train)
s = 0 # "общая" дисперсия
for i, component in enumerate(pca.components_):
  percent = round( 100 * pca.explained_variance_ratio_[i], 2 )
  print(f"{i + 1} component: {percent}% of initial variance")
  s += percent
  if i == 0:
  # для первой компоненты делаем словарь, у которого в качестве ключа
  # будет вес того или иного признака(т.е. индексируем по вкладу)
    d = dict()
    for value, name in zip(component,train.columns):
      d[abs(round(value, 2))] = name
  print(str(s)[:5], '\n')
max_weight = max(d.keys())
print(f"Признак с максимальным вкладом в 1-ую компоненту: {d[max_weight]}")
```

```python
# 16) Построить двухмерное представление данных с помощью алгоритма t-SNE
from sklearn.manifold import TSNE
import plotly.express as px
import matplotlib.pyplot as plt
embed = TSNE(n_components=2, random_state=42)
df_embedded = embed.fit_transform(df.select_dtypes(include=['number']))  # Только числовые признаки
print('Новая размерность данных df:', df_embedded.shape)
features = ["school", "sex", "age", "address", "famsize", "Pstatus", "Medu", "Fedu",
            "Mjob", "Fjob", "reason", "guardian", "traveltime", "studytime",
            "failures", "schoolsup", "famsup", "paid", "activities", "nursery",
            "higher", "internet", "romantic", "famrel", "freetime", "goout", "Dalc",
            "Walc", "health", "absences"]

plt.figure(figsize=(15, 30))
for i, feature in enumerate(features, 1):
    plt.subplot(10, 3, i)
    sns.scatterplot(x=df_embedded[:, 0], y=df_embedded[:, 1],
                   hue=df[feature], palette='viridis', alpha=0.7)
    plt.title(f't-SNE по признаку {feature}')
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
